### Questão 2:

Desenvolver um algoritmo que efetue a soma de todos os números ímpares que são múltiplos de três e que se encontram no conjunto dos números de 1 até 500.

```python
numero = 1
soma = 0

while numero <= 500:
  if numero%3 == 0 and numero%2 == 1:
    soma += numero
  numero += 1

print(soma)
```

### Questão 3:

Desenvolver um algoritmo que leia a altura de 15 pessoas. Este programa deverá calcular e mostrar :

a. A menor altura do grupo;

b. A maior altura do grupo;

```python
pessoas = 15
menor = 0
maior = 0
cont = 0

while pessoas > cont:
  altura = float(input("digite sua altura em metros: "))

  if cont == 0:
    maior = altura
    menor = altura
  else:
    if altura > maior:
      maior = altura
    elif altura < menor:
      menor = altura

  cont += 1

print(f"\na maior altura registrada foi de {maior}m e a menor for {menor}m")
```

### Questão 4:

Construa um programa que lê um número inteiro N a partir do teclado e calcula a raiz quinta inteira desse número, se houver. Caso não haja, ele deve informar isso ao usuário.




```python
running = True
while running:
  numero = int(input("digite um inteiro: "))
  raiz = numero ** (1/5)

  if int(raiz) == raiz:
    print(f"tem raiz inteira: {raiz}")
    running = False
  else:
    print(f"não tem raiz inteira: {raiz}")
    print("digite outro número")
```

### Questão 5:

Escrever um algoritmo que leia uma quantidade desconhecida de números e conte quantos deles estão nos seguintes intervalos: [0-25], [26-50], [51-75] e [76-100]. A entrada de dados deve terminar quando for lido um número negativo.

```python
numero = float(input("digite um numero positivo: "))
numeros_contidos = 0

if (0 <= numero <= 25) or (26 <= numero <= 50) or (51 <= numero <= 75) or (76 <= numero <= 100):
  numeros_contidos += 1

while numero >=0 :
  numero = float(input("digite um numero positivo: "))
  if (0 <= numero <= 25) or (26 <= numero <= 50) or (51 <= numero <= 75) or (76 <= numero <= 100):
    numeros_contidos += 1

print(f"foram digitados {numeros_contidos} numeros dentro do intervalo")
```

### Questão 6:

```python
numero = int(input("digite um numero positivo: "))

quant = 1
media_geral = numero
impares = 0
pares = 0
media_pares = 0

if numero > 0:
  if numero%2 == 0:
    pares += 1
    media_pares += numero
  else:
    impares += 1



while numero > 0 :
  numero = int(input("digite um numero positivo: "))
  quant += 1
  media_geral += numero

  if numero > 0:
    if numero%2 == 0:
      pares += 1
      media_pares += numero
    else:
      impares += 1

media_geral /= quant
media_pares /= pares

print(f"foram digitados {impares} numeros ímpares, {pares} pares (média de {media_pares}), e média geral de {media_geral}")
```
