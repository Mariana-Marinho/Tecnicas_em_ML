 1) Faça um programa que:


>  a) calcula o salário dos representantes de venda de uma empresa: o salário total será o salário base R$ 4200 mais 20 reais por cada venda (cuja quantidade está armazenada na variável qtdeVendas); Bloco com recuo

>  b) calcula o lucro total da empresa em uma semana, A quantidade de vendas pordia da semana estão nas variáveis qtdeVendasSegunda, qtdeVendasTerca, ...., qtdeVendasSabado, qtdeVendasDomingo. As vendas que acontecem em dia de semana custam 300 reais e as vendas que acontecem nos finais de semana custam 250 reais.

```python
qtdeVendasSegunda = int(input("vendas segunda: "))
qtdeVendasTerca = int(input("vendas terça: "))
qtdeVendasQuarta = int(input("vendas quarta: "))
qtdeVendasQuinta = int(input("vendas quinta: "))
qtdeVendasSexta = int(input("vendas sexta: "))
qtdeVendasSabado = int(input("vendas sabado: "))
qtdeVendasDomingo = int(input("vendas domingo: "))

salario_base = 4200
comissao = 20
salario_total = salario_base + (comissao * (qtdeVendasSegunda + qtdeVendasTerca + qtdeVendasQuarta + qtdeVendasQuinta +
                                           qtdeVendasSexta + qtdeVendasSabado + qtdeVendasDomingo))

valor_sabdom = 250
valor_segsex = 300
receita = (valor_segsex * (qtdeVendasSegunda + qtdeVendasTerca + qtdeVendasQuarta + qtdeVendasQuinta +
                          qtdeVendasSexta)) + (valor_sabdom * (qtdeVendasSabado + qtdeVendasDomingo))

lucro = receita - salario_total
print(f"\no lucro da empresa foi de {lucro}")
```

2) Faça um programa que tenha as variáveis *qtdeVacinaDisponivel*, *qtdeVacinados*, *qtdePopulacaoVacinavel* e *qtdeProfissionaisSaude*, e que gere cálculo dos seguintes valores:

> a) quantidade de pessoas que faltam ser vacinadas;

> b) quantidade de vacinas restantes se toda a população que ainda não se vacinou se vacine;

> c) quantos profissinais de saúde serão necessários para vacinar as pessoas que ainda não se vacinaram;

Considere que para vacinar 10 pessoas, é preciso em média 2 profissionais da saúde serão necessários para vacinar as pessoas que ainda não se vacinaram.




```python
qtdeVacinaDisponivel = int(input("quantas vacinas serão distribuídas: "))
qtdeVacinados = int(input("quantas pessoas foram vacinadas: "))
qtdePopulacoVacinavel = int(input("quantas pessoas são vacináveis: "))
qtdeProfissionaisSaude = int(input("quantos profissionais da saude: "))

nao_vacinados = qtdePopulacoVacinavel - qtdeVacinados
vacinas_restantes = qtdeVacinaDisponivel - qtdeVacinados

profissionais_necessarios = ((nao_vacinados//10)+(nao_vacinados%10 > 0))
profissionais_necessarios *= 2

print(f"\n{nao_vacinados} pessoas ainda não se vacinaram")
print(f"{vacinas_restantes} vacinas ainda faltam ser aplicadas")
print(f"{profissionais_necessarios} profissionais serão necessários para vacinar essas pessoas")
```
